###Direct wave

\begin{equation}
t_1\,=\,\frac{x}{V_1}
\end{equation}

###First head wave

\begin{equation}
t_2\,=\,\frac{x}{V_2}\,+\,\frac{2h_1\cos(i_{12})}{V_1}
\,=\,\frac{x}{V_2}\,+\frac{2h_1\sqrt{{V_2}^2\,-\,{V_1}^2}}{V_1V_2}
\end{equation}

###Second head wave

\begin{equation}
t_3\,=\,\frac{x}{V_3}
   \,+\,\frac{2h_1\cos(i_{13})}{V_1}
   \,+\,\frac{2h_2\cos(i_{23})}{V_2}
\end{equation}
<br>
\begin{equation}
\sin(i_{mn})\,=\,\frac{V_m}{V_n}
\implies \cos(i_{mn})\,=\,\sqrt{1\,-\,{\frac{V_m}{V_n}}^2}
\,=\,\frac{\sqrt{{V_n}^2\,-\,{V_m}^2}}{V_n}
\end{equation}
<br>
\begin{equation}
t_3\,=\,\frac{x}{V_3}
   \,+\,\frac{2h_1\sqrt{{V_3}^2\,-\,{V_1}^2}}{V_1V_3}
   \,+\,\frac{2h_2\sqrt{{V_3}^2\,-\,{V_2}^2}}{V_2V_3}
\end{equation}
<br>

```python
import math
import numpy as np
```

```python
def directWave (x, v1):

  return x / v1

def headWave1 (x, v1, v2, h1):

  return x / v2 + 2 * h1 * math.sqrt (v2 ** 2 - v1 ** 2) / (v1 * v2)

def headWave2 (x, v1, v2, v3, h1, h2):

  return x / v3 + 2 * h1 * math.sqrt (v3 ** 2 - v1 ** 2) / (v1 * v3) +\
                  2 * h2 * math.sqrt (v3 ** 2 - v2 ** 2) / (v2 * v3)
```

```python
if __name__ == '__main__':

# Modify the parameters and re-execute the code to see
# how the travel-time curves change

  xmin = 0   # (km)
  xmax = 40  # (km)
  h1   = 2.0 # (km)
  h2   = 3.0 # (km)
  v1   = 1.7 # (km/s)
  v2   = 4.0 # (km/s)
  v3   = 6.0 # (km/s)
```

```python
if __name__ == '__main__':

  plt.figure (figsize = (16, 8), dpi = 100)

  x   = np.linspace (xmin, xmax)
  t1  = directWave (x, v1)
  tr1 = headWave1 (x, v1, v2, h1)
  tr2 = headWave2 (x, v1, v2, v3, h1, h2)

  plt.plot (x, t1, color = 'red',
            linewidth = 2,
            label = 'Direct wave')
  plt.plot (x, tr1, color = 'green',
            linewidth = 2,
            label = 'Head wave 1')
  plt.plot (x, tr2, color = 'blue',
            linewidth = 2,
            label = 'Head wave 2')

  plt.title ('Travel-time curves',
             fontsize = 28,
             y = 1.05)
  plt.xlabel ('Distance from the source (km)',
              fontsize = 22,
              labelpad = 10)
  plt.ylabel ('Travel time (s)',
              fontsize = 22,
              labelpad = 10)
  plt.tick_params (axis = 'x', labelsize = 16)
  plt.tick_params (axis = 'y', labelsize = 16)
  plt.legend (loc = 2, fontsize = 18)
  plt.grid ()
  plt.show ()
```
